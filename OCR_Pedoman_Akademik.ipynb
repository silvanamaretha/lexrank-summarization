```python
import pytesseract
from pdf2image import convert_from_path
import os
```

```python
# === 1. Upload file PDF ===
from google.colab import files
uploaded = files.upload()

pdf_path = list(uploaded.keys())[0]
```

```python
# === 2. Convert PDF ke gambar per halaman ===
pages = convert_from_path(pdf_path, dpi=300)

# Buat folder simpan gambar
os.makedirs("pages", exist_ok=True)

texts = []
```

```python
# === 3. OCR tiap halaman ===
for i, page in enumerate(pages):
    image_path = f"pages/page_{i+1}.jpg"
    page.save(image_path, "JPEG")

    text = pytesseract.image_to_string(image_path, lang="ind")  # pakai bahasa Indonesia
    texts.append(text)

    print(f"OCR halaman {i+1} selesai...")
```

```python
# === 4. Simpan hasil OCR ke file TXT ===
output_file = "hasil_ocr.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write("\n\n".join(texts))

print("OCR selesai ✅. File hasil: ", output_file)
```

```python
# === 5. Download hasil OCR ===
files.download(output_file)
```
